```python
import os
import glob
import tqdm
import pathlib

import cobra
import json

from mapper import *
from maps import * 
from reader import *
from merger import *
```

```python
#MAPS = ["maps/gly.xml", "maps/tca.xml", "maps/ppp.xml", "maps/glycoxylate.xml"]
MAPS = ["maps copy/00010.xml", "maps copy/00020.xml", "maps copy/00030.xml"]
MODELS = glob.glob("test_model/*.xml")

OUTPUT = "output_maps_test/"

MODELS
```

```python
for model_path in tqdm.tqdm(MODELS):

    name = list(os.path.basename(model_path).split("."))[0]
    output_path = OUTPUT+name
    pathlib.Path(output_path).mkdir(parents=True, exist_ok=True)

    merger = EscherMerger()
    merger.reset()

    for map in MAPS:

        m = KeggMap()
        with open(map, "r") as f:
            m.read_from_file(f)

        model = cobra.io.read_sbml_model(model_path)
        mapper = EscherMapper(model)

        escher_map = mapper.map_flux_on_kegg(m)

        merger.append(escher_map)

        map_output_path = output_path + "/" + os.path.basename(map.replace("xml", "json"))

        with open(map_output_path, "w") as f:
            json.dump(escher_map, f)
            print(map_output_path, " is saved!")


    print(len(merger.maps))
    merged = merger.merge_maps()

    merged_output_path = output_path + "/merged_" + name + ".json"

    with open(merged_output_path, "w") as f:
        json.dump(merged, f)
        print(merged_output_path, " is saved!")
```
