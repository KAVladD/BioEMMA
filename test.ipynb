```python
import json
import os

import escher
import cobra

from mapper_base import EscherMapper
from maps import KeggMap
```

```python
m = KeggMap()
with open("temp.xml", "r") as f:
    m.read_from_file(f)
# m.read_from_url("https://rest.kegg.jp/get/rn00010/kgml")
```

```python
metabolites = m.get_metabolites()
reactions = m.get_reactions()
```

```python
e = EscherMapper(metabolites, reactions, markers_dist=10, scaling_factor=6, database="SEED", axis_epsilon=10, remove_orphan_metabolites=True)
```

```python
cobra_model = cobra.io.read_sbml_model("test_model/agita_genomic_modified.xml")
```

```python
if os.path.exists("output.json"):
    os.remove("output.json")

emap = e.build_map(cobra_model)

with open("output.json", "w") as f:
    json.dump(emap, f)
    print("output.json is saved!")
```

```python
if os.path.exists("output_kegg.json"):
    os.remove("output_kegg.json")

emap = e.build_kegg_map()

with open("output_kegg.json", "w") as f:
    json.dump(emap, f)
    print("output_kegg.json is saved!")
```
