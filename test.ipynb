```python
import json
import os

import escher

from mapper_base import EscherMapper
from maps import KeggMap
```

```python
m = KeggMap()
with open("temp.xml", "r") as f:
    m.read_from_file(f)
# m.read_from_url("https://rest.kegg.jp/get/rn00010/kgml")
```

```python
metabolites = m.get_metabolites()
reactions = m.get_reactions()
```

```python
reactions
```

```python
e = EscherMapper(metabolites, reactions)
```

```python
if os.path.exists("output.json"):
    os.remove("output.json")

emap = e.build_map()

with open("output.json", "w") as f:
    json.dump(emap, f)
    print("output.json is saved!")
```

```python
builder = escher.Builder(map_json="output.json")
```
